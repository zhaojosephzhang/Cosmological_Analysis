# Ray Tracing Without Box Rotation

This notebook runs the **box-stacking / ray-tracing** light-cone method without rotating the simulation box. A single observer is placed inside the periodic volume, and the sightlines are propagated through the stacked snapshots by periodic wrapping only.

## Method Summary

The core production code is stored in [`Ray_tracing_no_roating.py`](./Ray_tracing_no_roating.py).

Key points of this version:
- it preserves the native box orientation and therefore does **not** randomize structures between shells
- it can sample either the full sky or the **CHIME declination window**
- when `use_chime_window=True`, the directions are drawn with `dec_min_deg <= declination <= dec_max_deg`
- snapshot transitions are smoothed with a short overlap region rather than a hard shell boundary

```python
from Ray_tracing_no_roating import RayConeConfig, build_dm_cone, save_dm_cone_hdf5, default_output_path
```

## Configuration and Cone Construction

This panel defines the snapshot range, the LoS sampling, and whether the CHIME sky selection is applied.

```python
cfg = RayConeConfig(
    base_dir='/home/zhaozhang/local/Grid_data/L100N1024_NoBH/',
    agn_tag='n',
    level_input='7',
    snap_min=15,
    snap_max=20,
    n_dir=100000,
    z_max=1.0,
    overlap_bins=16,
    n_out=300,
    use_chime_window=True,
    dec_min_deg=-9.5,
    dec_max_deg=90.0,
    verbose=True,
)

dm_tot, dm_cum, z_prof, n_hat, meta = build_dm_cone(cfg)
print(dm_tot.shape, dm_cum.shape, z_prof.shape, n_hat.shape)
```

## Save the Data Product

Only the data-generation part is retained here. The output HDF5 file stores the total DM, the cumulative `DM(<z)` profiles, the sky directions, the snapshot list, and the CHIME-window metadata.

```python
out_path = default_output_path(cfg)
save_dm_cone_hdf5(out_path, cfg, dm_tot, dm_cum, z_prof, n_hat, meta)
print(out_path)
```
