# Ray Tracing With Rotating Boxes

This notebook runs the **box-stacking / ray-tracing** light-cone method in the rotating-box mode. The observer remains fixed, but each radial shell receives a shared random box transform.

## What Changes Relative to the Non-Rotating Version

The only required conceptual change in `rotating_box` is the following:
- when the light cone moves from one shell to the next, the box orientation is randomly rotated, and optionally translated, to reduce repeated structures along the line of sight
- the same shell transform is shared by **all** rays in that shell so that the angular correlation inside the shell is not artificially erased

The core production code is stored in [`Ray_tracing_rotating_box.py`](./Ray_tracing_rotating_box.py).

```python
from Ray_tracing_rotating_box import RayConeConfig, build_dm_cone, save_dm_cone_hdf5, default_output_path
```

## Configuration and Cone Construction

This panel controls the shell rotation / translation flags as well as the radial sampling used to generate the cumulative DM profiles.

```python
cfg = RayConeConfig(
    base_dir='/home/zhaozhang/local/Grid_data/L100N1024_Fiducial/',
    agn_tag='f',
    level_input='8',
    snap_min=15,
    snap_max=20,
    n_dir=30000,
    z_max=1.0,
    rotate_each_shell=True,
    translate_each_shell=True,
    n_profile=300,
    step_mult=1.0,
    overlap_bins=16,
    verbose=True,
)

dm_tot, dm_cum, z_prof, n_hat, meta = build_dm_cone(cfg)
print(dm_tot.shape, dm_cum.shape, z_prof.shape, n_hat.shape)
```

## Save the Data Product

This notebook keeps only the core data-production path. The HDF5 output records the cumulative DM cone together with the metadata required to reproduce the shell-rotation setup.

```python
out_path = default_output_path(cfg)
save_dm_cone_hdf5(out_path, cfg, dm_tot, dm_cum, z_prof, n_hat, meta)
print(out_path)
```
