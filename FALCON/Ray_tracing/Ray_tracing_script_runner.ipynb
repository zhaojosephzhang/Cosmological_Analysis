# Ray Tracing Script Runner

This notebook is the lightweight driver version. It directly imports the existing Python scripts, sets a configuration, runs the cone builder, and writes the HDF5 output.

## What This Notebook Is For

- keep the heavy implementation inside the `.py` scripts
- keep the notebook as a clean control panel for configuration and execution
- make it easy to switch between the non-rotating and rotating-box pipelines
- keep the CHIME-window option explicit in the non-rotating case

```python
from Ray_tracing_no_roating import (
    RayConeConfig as NoRotConfig,
    build_dm_cone as build_dm_cone_no_rot,
    save_dm_cone_hdf5 as save_dm_cone_hdf5_no_rot,
    default_output_path as default_output_path_no_rot,
)

from Ray_tracing_rotating_box import (
    RayConeConfig as RotConfig,
    build_dm_cone as build_dm_cone_rot,
    save_dm_cone_hdf5 as save_dm_cone_hdf5_rot,
    default_output_path as default_output_path_rot,
)
```

## Non-Rotating Cone With Optional CHIME Window

Set `use_chime_window=True` if you want the LoS sampling to follow the CHIME declination coverage.

```python
cfg_no_rot = NoRotConfig(
    base_dir='/home/zhaozhang/local/Grid_data/L100N1024_NoBH/',
    agn_tag='n',
    level_input='7',
    snap_min=15,
    snap_max=20,
    n_dir=100000,
    z_max=1.0,
    overlap_bins=16,
    n_out=300,
    use_chime_window=True,
    dec_min_deg=-9.5,
    dec_max_deg=90.0,
    verbose=True,
)

dm_tot_no_rot, dm_cum_no_rot, z_prof_no_rot, n_hat_no_rot, meta_no_rot = build_dm_cone_no_rot(cfg_no_rot)
print(dm_tot_no_rot.shape, dm_cum_no_rot.shape, z_prof_no_rot.shape, n_hat_no_rot.shape)
```

```python
out_path_no_rot = default_output_path_no_rot(cfg_no_rot)
save_dm_cone_hdf5_no_rot(
    out_path_no_rot,
    cfg_no_rot,
    dm_tot_no_rot,
    dm_cum_no_rot,
    z_prof_no_rot,
    n_hat_no_rot,
    meta_no_rot,
)
print(out_path_no_rot)
```

## Rotating-Box Cone

This panel runs the shell-by-shell rotating-box version. The main difference is that each shell can carry a shared random rotation and translation.

```python
cfg_rot = RotConfig(
    base_dir='/home/zhaozhang/local/Grid_data/L100N1024_Fiducial/',
    agn_tag='f',
    level_input='8',
    snap_min=15,
    snap_max=20,
    n_dir=30000,
    z_max=1.0,
    rotate_each_shell=True,
    translate_each_shell=True,
    n_profile=300,
    step_mult=1.0,
    overlap_bins=16,
    verbose=True,
)

dm_tot_rot, dm_cum_rot, z_prof_rot, n_hat_rot, meta_rot = build_dm_cone_rot(cfg_rot)
print(dm_tot_rot.shape, dm_cum_rot.shape, z_prof_rot.shape, n_hat_rot.shape)
```

```python
out_path_rot = default_output_path_rot(cfg_rot)
save_dm_cone_hdf5_rot(
    out_path_rot,
    cfg_rot,
    dm_tot_rot,
    dm_cum_rot,
    z_prof_rot,
    n_hat_rot,
    meta_rot,
)
print(out_path_rot)
```
